# Run Synthetic Problem Instances and Generate Plots of Fraction Solved

```julia
include("tests/tests.jl")
Hrand = runSuiteOfTest("tests/rand_test_problems",methods, [1e-3, 1e-6, 1e-9], 500)
```

```julia
p1 = plot(title = "Low Accuracy On Random Instances", legend=:outertopright, xlabel="Iterations", ylabel="Fraction Instances Solved", ylims=(0,1))
colors = [palette(:tab10, 6)[1] palette(:tab10, 6)[2] palette(:tab10, 6)[3] palette(:tab10, 6)[4] palette(:tab10, 6)[5] palette(:tab10, 6)[6]]
labels = ["OGM", "SPGM-2", "SPGM-10", "SPGM", "BFGS", "LBFGS-10"]
for m in 1:5
    plot!(Hrand[m,1,1:40]/42.0, linewidth=3, lc = colors[m], label = labels[m])
end
savefig(p1, "lowAccur_rand.png")
p1
```

```julia
p1 = plot(title = "Medium Accuracy On Random Instances", legend=:outertopright, xlabel="Iterations", ylabel="Fraction Instances Solved", ylims=(0,1))
colors = [palette(:tab10, 6)[1] palette(:tab10, 6)[2] palette(:tab10, 6)[3] palette(:tab10, 6)[4] palette(:tab10, 6)[5] palette(:tab10, 6)[6]]
labels = ["OGM", "SPGM-2", "SPGM-10", "SPGM", "BFGS", "LBFGS-10"]
for m in 1:5
    plot!(Hrand[m,2,:]/42.0, linewidth=3, lc = colors[m], label = labels[m])
end
savefig(p1, "medAccur_rand.png")
p1
```

```julia
p1 = plot(title = "High Accuracy On Random Instances", legend=:outertopright, xlabel="Iterations", ylabel="Fraction Instances Solved", ylims=(0,1))
colors = [palette(:tab10, 6)[1] palette(:tab10, 6)[2] palette(:tab10, 6)[3] palette(:tab10, 6)[4] palette(:tab10, 6)[5] palette(:tab10, 6)[6]]
labels = ["OGM", "SPGM-2", "SPGM-10", "SPGM", "BFGS", "LBFGS-10"]
for m in 1:5
    plot!(Hrand[m,3,:]/42.0, linewidth=3, lc = colors[m], label = labels[m])
end
savefig(p1, "highAccur_rand.png")
p1
```
