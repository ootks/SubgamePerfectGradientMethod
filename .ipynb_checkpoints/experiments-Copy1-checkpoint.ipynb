```julia
#Code below reads data from the libsvm database and populates the associated logistic regression/classifcation problem

using HTTP

dataSets = ["australian","diabetes","heart","ionosphere_scale","splice","sonar_scale"]

for s in dataSets
    # URL of the dataset
    url = "https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/binary/"*s

    # Fetch the data from the URL
    response = HTTP.get(url)
    data_lines = split(String(response.body), '\n') |> filter(x -> !isempty(x))  # Split into lines and remove empty lines

    # Parse the data to find the maximum index
    max_index = 0
    for line in data_lines
        tokens = split(line)
        for token in tokens[2:end]
            idx, _ = split(token, ":")
            max_index = max(max_index, parse(Int, idx))
        end
    end

    # Initialize vector b and matrix A
    b = Float64[]
    rows = []

    for line in data_lines
        tokens = split(line)  # Split the line by whitespace
        push!(b, parse(Float64, tokens[1]))  # First number goes to b

        # Create a row of size max_index initialized to zeros
        row = zeros(Float64, max_index)
        for token in tokens[2:end]
            idx, val = split(token, ":")
            row[parse(Int, idx)] = parse(Float64, val)
        end
        push!(rows, row)
    end

    # Convert rows into a matrix
    A = hcat(rows...) |> transpose  # Transpose to ensure proper orientation
        
    A = A/norm(A)
    (m,n) = size(A)
    x0= zeros(n)
    
        
    title = "LogisticRegression_"*s*"_dim"*string(n)
    description = "Least Squares Regression minimizing (1/m)sum log(1+exp(bi ai^Tx)) + 1/2 ||x||^2 with data pulled from LIBSVM "*s*"and x0=0, n="*string(n)*", m="*string(m)

    q = SumOracle(LinearCompositionOracle(logisticRegression(b), A, zeros(m)), quadratic(Matrix{Float64}(I, n, n),zeros(n),0))

    errors, guarantees, xStar = test( SPGM(100), q, x0, 2000, 0)
    e2, guarantees, xStar = test( BFGS(), q, x0, 2000, 0)
    e3, guarantees, xStar = test( OGM(), q, x0, 2000, 0)
    minVal = min(min(minimum(errors),minimum(e2)),minimum(e3))

    @save "tests/real_test_problems/"*title*".jld2" title description q x0 minVal xStar
end
```

```julia
#Code below reads data from the libsvm database and populates the associated least squares regression problem

using HTTP

dataSets = ["housing","mpg","pyrim","space_ga","triazines","bodyfat"]

for s in dataSets
    # URL of the dataset
    url = "https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/regression/"*s

    # Fetch the data from the URL
    response = HTTP.get(url)
    data_lines = split(String(response.body), '\n') |> filter(x -> !isempty(x))  # Split into lines and remove empty lines

    # Parse the data to find the maximum index
    max_index = 0
    for line in data_lines
        tokens = split(line)
        for token in tokens[2:end]
            idx, _ = split(token, ":")
            max_index = max(max_index, parse(Int, idx))
        end
    end

    # Initialize vector b and matrix A
    b = Float64[]
    rows = []

    for line in data_lines
        tokens = split(line)  # Split the line by whitespace
        push!(b, parse(Float64, tokens[1]))  # First number goes to b

        # Create a row of size max_index initialized to zeros
        row = zeros(Float64, max_index)
        for token in tokens[2:end]
            idx, val = split(token, ":")
            row[parse(Int, idx)] = parse(Float64, val)
        end
        push!(rows, row)
    end

    # Convert rows into a matrix
    A = hcat(rows...) |> transpose  # Transpose to ensure proper orientation
    (m,n) = size(A)
    x0= zeros(n)
        
    title = "LSRegression_"*s*"_dim"*string(n)
    description = "Least Squares Regression minimizing (1/m)||Ax-b||^2_2 regularized by sum of Huber of x_i with data pulled from LIBSVM "*s*"and x0=0, n="*string(n)*", m="*string(m)*", L=100, r=1"
        
    q =  SumOracle(LinearCompositionOracle(quadratic(Matrix{Float64}(I, m, m),zeros(m),0),A/sqrt(m/2),b/sqrt(m/2)), huberL1(100,1))

    errors, guarantees, xStar = test( SPGM(100), q, x0, 2000, 0)
    e2, guarantees, xStar = test( BFGS(), q, x0, 2000, 0)
    e3, guarantees, xStar = test( OGM(), q, x0, 2000, 0)
    minVal = min(min(minimum(errors),minimum(e2)),minimum(e3))
        
    @save "tests/real_test_problems/"*title*".jld2" title description q x0 minVal xStar
end
```
